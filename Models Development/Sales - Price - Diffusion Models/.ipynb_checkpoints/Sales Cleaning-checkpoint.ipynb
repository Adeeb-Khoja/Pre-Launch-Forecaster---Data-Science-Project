<h1>Cleaning<h1>

```python
import pandas as pd
```

```python
pd.options.display.max_rows = 999
df = pd.read_csv("All-Sales.csv")
df.head()
```

```python
df.shape
# 15 columns , 157 rows
```

```python
df.dtypes
#change price, resale value to float
```

```python
# Unique values
df.nunique()
#30 manufacturer and 156 different car models
#
```

```python
# Missing Values
df.isna().sum()
# Missing values represented as . in this dataset
```

```python
# Checking duplicates
df[df.duplicated(subset='Model',keep= False)].sort_values('Model')
# the duplicates are of different manufacturer
# It is better to combine Model with manufacturer
```

```python
#### Combining Model and Manufacturer
import numpy as np
Manuf = list(df['Manufacturer'])
models = list(df['Model'])
combined = []
for i in range(len(models)):
    combined = np.append(combined, str(Manuf[i])+"- "+str(models[i]) )

df['Manf/Model'] = combined
df.head()
```

```python
# Dropping uncecessary columns
df_sales = df.drop(df.columns[[0,1,3,4,6,7,8,9,10,11,12,13,]],axis=1)
df_sales.head()
```

```python
# Reordering columns
df_sales = df_sales[df_sales.columns[[3,0,1,2]]]
df_sales.head()
```

```python
# Dropping missing values as .
df_sales.drop(df_sales[df_sales['Price in thousands']=="."].index,axis=0,inplace=True)
df_sales.drop(df_sales[df_sales['Sales in thousands']=="."].index,axis=0,inplace=True)
df_sales
```

```python
df_sales.shape
# 2 prices were missing and dropped
```

```python
# tranform date column to date data type
df_sales['Latest Launch'] = pd.to_datetime(df_sales['Latest Launch'], format='%d-%b-%y')
df_sales.head()
```

```python
# Renaming columns
new = {'Manf/Model':'Model', 'Sales in thousands':'Sales','Price in thousands':'Price',
       'Latest Launch':'Launch-Date'}
df_sales.rename(columns=new,inplace=True)
df_sales.head()
```

```python
# New column to count the number of days form launch till the data collected (23-10-2017)
df_sales['Days'] = ( pd.Timestamp(2017, 10, 23) - df_sales['Launch-Date']).astype(str)
for i in df_sales.index:
    df_sales.loc[i, 'Days'] = float(df_sales.loc[i, 'Days'].split()[0])
df_sales['Months'] = (df_sales['Days']/30).astype(int)
df_sales['Years'] = (df_sales['Days']/365).astype(int)
df_sales.head()
```

```python
# Change data types of price and sales
df_sales['Sales'].astype(float)
df_sales['Price'] = pd.to_numeric(df_sales['Price'])
df_sales.dtypes
```

```python
# Saving the cleand data set
df_sales.to_csv('cleaned-sales.csv',index=False)
```
